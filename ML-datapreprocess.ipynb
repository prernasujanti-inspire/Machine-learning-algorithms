```python
# Data Preprocessing

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Handling Categorical Variables
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# Importing the dataset
dataset = pd.read_csv('sample_data.csv')
```

```python
dataset.head()
```

```python
# Encoding the size column  (Ordinal Encoding)

dataset['Size'] = dataset['Size'].map({'S': 1, 'M': 2, 'L': 3})
```

```python
dataset.head()
```

```python
# Encoding the colour column (One Hot Encoding) for nominal data variables
# using get_dummies() method of pandas

#dataset2 = pd.get_dummies(dataset, columns=['Colour'])  both are same
dataset2 = pd.get_dummies(dataset[['Colour', 'Size', 'Price','Qty']])
dataset2.head()
```

```python
# the last column has multicollinearity problem as it is highly correlated with the 2 colour columns, so we can drop it to avoid the dummy variable trap

dataset2 = pd.get_dummies(dataset[['Colour', 'Size', 'Price','Qty']], drop_first=True)
dataset2.head()
```

```python
# other encoding techniques for categorical variables are Label Encoding, Binary Encoding, Target Guided Encoding, etc.
```

```python
dataset2.columns
```

```python
# Scaling the features
# Standardization (Z-score Normalization) is a common technique for feature scaling. It transforms the data to have a mean of 0 and a standard deviation of 1.

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
dataset2_scaled = dataset2.copy()
dataset2_scaled[['Size','Price','Qty']] = sc.fit_transform(dataset2[['Size','Price','Qty']])
dataset2_scaled
```

```python
# standard deviation is 1 and mean is 0 after standardization
print('Mean:', dataset2_scaled.mean())
print('Standard Deviation:', dataset2_scaled.std())
```

```python
# standard deviation of 2,6,10)

a = [2, 6, 10]
print('Mean:', np.mean(a))
print('Standard Deviation:', np.std(a))
```

```python
# after scaling the original data is lost, so we can convert the scaled data back to a dataframe and assign the column names to it

dataset2_scaled_df = pd.DataFrame(dataset2_scaled, columns=dataset2.columns)
dataset2_scaled_df.head()
```

```python
# after fitting the model, we can use the inverse_transform() method to get back the original data from the scaled data

dataset2_original = sc.inverse_transform(dataset2_scaled)
dataset2_original_df = pd.DataFrame(dataset2_original, columns=dataset2.columns)
```
