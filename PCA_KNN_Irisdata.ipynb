```python
# Check PCA effect on a dataset and on a KNN model.
# take the iris dataset
# load needed libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import load_iris
```

```python
# split the data into training and testing sets
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
# Visualize the data before PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', edgecolor='k', s=100)
plt.title('Iris Dataset - Before PCA')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# PCA algorithm
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# visualize the data after PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='viridis', edgecolor='k', s=100)
plt.title('Iris Dataset - After PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# which variable can be removed after PCA?
# PCA creates new features (principal components) that are linear combinations of the original features.
# After applying PCA, the original features can be removed if you only want to use the principal components for further analysis or modeling.
# K Nearest Neighbors (KNN) algorithm
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, p=2)  # Using Euclidean distance
knn.fit(X_train_pca, y_train)
accuracy = knn.score(X_test_pca, y_test) * 100
print(f'KNN Accuracy after PCA: {accuracy:.2f}%')
```

```python
# Summary
# For Iris dataset,
# KNN accuracy before PCA and without hyper parameter tuning: 96.6%
# KNN accuracy before PCA and with hyper parameter tuning: 93%
# KNN accuracy after PCA and with hyper parameter tuning: 100%

# Accuracy increases by applying PCA
```
